```python
import spacy
spacy.load('en')
from spacy.lang.en import English
parser = English()

def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
for w in ['dogs', 'ran', 'discouraged']:
    print(w, get_lemma(w), get_lemma2(w))
```

```python
nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
import random
text_data = []
with open('titles.csv') as f:
    for line in f:
        tokens = prepare_text_for_lda(line)
#         if random.random() > .99:
        print(tokens)
        text_data.append(tokens)
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
```

```python
corpus = [dictionary.doc2bow(text) for text in text_data]
```

```python
import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
import re
import numpy
from gensim.parsing.preprocessing import preprocess_string, strip_punctuation,strip_numeric

def getTopicForText(lda, question, numwords):
    temp = question.lower()
#     for i in range(len(punctuation_string)):
#         temp = temp.replace(punctuation_string[i], '')

    words = re.findall(r'\w+', temp, flags = re.UNICODE ) # | re.LOCALE)

    important_words = []
    important_words = filter(lambda x: x, words) # not in stoplist, words)

    dictionary = corpora.Dictionary.load('dictionary.gensim')

    ques_vec = []
    ques_vec = dictionary.doc2bow(important_words)

    topic_vec = []
    topic_vec = lda[ques_vec]

    word_count_array = numpy.empty((len(topic_vec), 2), dtype = numpy.object)
    for i in range(len(topic_vec)):
        word_count_array[i, 0] = topic_vec[i][0]
        word_count_array[i, 1] = topic_vec[i][1]

    idx = numpy.argsort(word_count_array[:, 1])
    idx = idx[::-1]
    word_count_array = word_count_array[idx]

    final = []
    final = lda.print_topic(word_count_array[0, 0], numwords)

#     question_topic = final.split('+') ## as format is like "probability * topic"
#     question_topic = question_topic.split('*') ## as format is like "probability * topic"

    # Get the words
    filters = [lambda x: x.lower(), strip_punctuation, strip_numeric]

    question_topic = preprocess_string(final, filters)

    return question_topic
```

```python
import pandas as pd
df = pd.read_csv('titles.csv')
df['Tags'] = pd.Series()
for i in range(len(df['title'])):
    df.loc[i, 'Tags'] = ','.join(getTopicForText(ldamodel, df['title'][i], numwords=4))
```
