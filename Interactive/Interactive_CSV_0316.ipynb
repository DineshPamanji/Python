# Interactive CSV 

## Set path

```python
# Set path
path = 'C:/Projects/FSOx/Data/Input/Data_Testing/'
```

```python
# Import packages
import os
import pandas as pd
import qgrid
#import itables.interactive
import ipywidgets as widgets

#!pip install qgrid

# Set display options
#qgrid.set_grid_option('forceFitColumns', False)
pd.options.display.max_columns = 25
pd.set_option('max_colwidth', 600)

filename = widgets.Dropdown(
    options=os.listdir(path),
    #value=2,
    description='Filename:',
    disabled=False,
)
filename
```

## Read input and display for filtering 

```python
# Read the input file
data = pd.read_csv(path+filename.value)

# Convert to qgrid widget - interactive table
qgrid_widget = qgrid.show_grid(data, show_toolbar=True)

qgrid_widget
```

## Save edited table

```python
# Save to the file in outputs
new_Jobs = qgrid_widget.get_changed_df()
new_Jobs.reset_index(inplace=True, drop=True)
```
